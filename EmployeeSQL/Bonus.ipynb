```python
from sqlalchemy import create_engine
import pandas as pd
from config import password
import matplotlib.pyplot as plt
```

```python
engine = create_engine(f'postgresql://harry:{password}@localhost:5432/employeesdb')
conn = engine.connect()
```

```python
salaryDF = pd.read_sql('SELECT * FROM salaries', conn)
salaryDF['salary'].hist(figsize = (20,8))
```

```python
# Query data
query = '''
SELECT title, AVG(salary) FROM employees
    JOIN titles ON titles.title_id = employees.emp_title_id
        JOIN salaries ON salaries.emp_no = employees.emp_no
GROUP BY title
'''

# Load into pandas
titleDF = pd.read_sql(query, conn)

# Plot
titleDF.plot('title', 'avg', kind = 'bar', figsize = (20,8), legend = False)
plt.xticks(rotation = 45, fontsize = 15)
plt.xlabel('Title', fontsize = 18)
plt.ylabel('Salary ($)', fontsize = 18)
plt.show()
```
